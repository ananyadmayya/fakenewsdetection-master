```python
import datetime
import spacy
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import pickle
from gensim.models import Word2Vec
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import MultinomialNB
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics import f1_score, classification_report, ConfusionMatrixDisplay, accuracy_score
```

```python
# load the preprocessed data
articledf = pd.read_csv('../data/article_preprocessed.csv')
```

```python
# define the features and labels
texts = articledf['tokens'].tolist()
labels = articledf['target'].values
```

```python
texts
```

```python
# create a custom tokenizer to feed to the TF-IDF Vectorizer
nlp = spacy.blank('en')

def custom_tokenizer(doc):
    return [t.text for t in nlp(doc) if not t.is_punct and not t.is_space and t.is_alpha]
```

```python
# Use the TF-IDF Vectorizer to create a simple embedding without saving the semantic and contextual meaning
vectorizer = TfidfVectorizer(tokenizer=custom_tokenizer)

# Fit and transform train data
X_train = vectorizer.fit_transform(texts)
y_train = labels

# Transform test data
X_test = vectorizer.transform(texts)
y_test = labels
```

```python

# Define model
clf = MultinomialNB()

# Train model
clf.fit(X_train, y_train)
```

```python
# Predict on test set
test_preds = clf.predict(X_test)

# Measure f1-score
print('Test set F1-score:', f1_score(y_test, test_preds, average='macro'))
```

```python
# measure accuracy score
accuracy_score(y_test, test_preds)
```
